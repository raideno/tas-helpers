```python
import numpy as np
import matplotlib.pyplot as plt

from tas_helpers.utils import generate_random_segmentation
from tas_helpers.visualization import SegmentationVisualizer

def main():
    labels = ["label1", "label2", "label3"]
    
    visualizer = SegmentationVisualizer(
        labels_values=labels,
    )
    
    segment = generate_random_segmentation(
        length=2000,
        labels=labels
    )

    visualizer.plot_segmentation(
        segment, 
        header="Example Video Segmentation", 
        fps=30,
        show_ticks=False
    )
    
    plt.show()
    
    print("[completed].")
    
if __name__ == "__main__":
    main()
```
